```python
import numpy as np
import pandas as pd

df = pd.read_csv("종관기상관측(2019)_0119.csv",encoding = 
                 'euc-kr',index_col = 0)
```

```python
df["일강수량(mm)"].fillna(0,inplace =True)
```

```python
df=df.dropna(axis = 0 )
```

```python
df.head()
```

```python
X, y = df.iloc[:,2:-1],df.iloc[:,-1]
```

```python
X.shape
```

```python
x_cen = X - X.mean(axis = 0)
x_cov = np.dot(x_cen.T , x_cen) / 363
print(x_cov)
```

```python
# 고유값 분해
w ,v =np.linalg.eig(x_cov)
print(w)
print("----------------",v)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca.fit(X)
PCA(copy=True, iterated_power='auto', n_components=3, random_state=None,
  svd_solver='auto', tol=0.0, whiten=False)
print('singular value :', pca.singular_values_)
print('singular vector :\n', pca.components_.T)
```

```python
pca.singular_values_
```

```python
print('eigen_value :', pca.explained_variance_)
print('explained variance ratio :', pca.explained_variance_ratio_)
```

```python
# PC1에 projection
pc1 = v[:, 0]
proj1 = np.dot(X, pc1)

# PC2에 projection
pc2 = v[:, 1]
proj2 = np.dot(X, pc2)

# PC3에 projection
pc3 = v[:, 2]
proj3 = np.dot(X, pc3)

proj_list = np.array([proj1.var(), proj2.var(), proj3.var()])

print('variance(==eigenvalue) :', proj_list)
print('explained variance ratio :', proj_list / proj_list.sum())

```

```python
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
print('선택할 차원 수 :', d)
```

```python
pca = PCA(n_components=0.95)
X_proj = pca.fit_transform(X)
print('principal component vec :\n', pca.components_.T)

#https://excelsior-cjh.tistory.com/167
```
